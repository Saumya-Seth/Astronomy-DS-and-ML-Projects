```python
import numpy as np
import pandas as pd
import matplotlib as plt
```

```python
data = np.load('galaxy_catalogue.npy')
```

```python
df = pd.DataFrame(data)
```

```python
df = df.sample(frac=1)
```

```python
df.reset_index(drop=True, inplace=True)
```

```python
df["conc_u"] = df['petroR50_u']/df['petroR90_u']
df["conc_r"] = df['petroR50_r']/df['petroR90_r']
df["conc_z"] = df['petroR50_z']/df['petroR90_z']
```

```python
df= df.drop("petroR50_u", axis=1).drop("petroR50_r", axis=1).drop("petroR50_z", axis=1).drop("petroR90_u", axis=1).drop("petroR90_r", axis=1).drop("petroR90_z", axis=1)
```

```python
df.head()
```

```python
Y = df["class"]
X = df.drop("class", axis=1)
```

```python
from sklearn.preprocessing import LabelEncoder
label_encoder = LabelEncoder()
Y = label_encoder.fit_transform(Y)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, train_size=0.72)
```

## Decision Tree

```python
from sklearn.tree import DecisionTreeClassifier
clf = DecisionTreeClassifier()
clf.fit(X_train, Y_train)
Y_preds = clf.predict(X_test)
print(Y_preds[:20])
```

```python
clf.score(X_test, Y_test)
```

```python
from sklearn.model_selection import cross_val_score
cv_acc = cross_val_score(clf, X, Y, cv=10, scoring=None)
```

```python
cv_acc
```

```python
cv_acc = np.mean(cv_acc)
cv_acc
```

## Random Forest

```python
from sklearn.ensemble import RandomForestClassifier 
rfc = RandomForestClassifier()
rfc.fit(X_train, Y_train)
Y_preds = rfc.predict(X_test)
print(Y_preds[:20])
```

```python
rfc.score(X_test, Y_test)
```

```python
from sklearn.model_selection import cross_val_score
cv_acc = cross_val_score(rfc, X, Y, cv=10, scoring=None)
np.mean(cv_acc)
```
